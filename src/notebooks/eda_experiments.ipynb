```python
import pickle

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme()
```

# Scenario I

```python
with open(file="../experiments/scenarioI/scenario_info.pkl", mode="rb") as f:
    scenario_info = pickle.load(f)
```

```python
qc = scenario_info["circuit"]
optimal_obj = scenario_info["optimal_obj_value"]
target = scenario_info["target"]
ising = scenario_info["ising"]
offset = scenario_info["offset"]
```

```python
n_shots = 4096
df_cobyla = pd.read_pickle(filepath_or_buffer=f"../experiments/scenarioI/results/scipy/COBYLA_n_shots_{n_shots}_experiments.pkl")
df_cmaes = pd.read_pickle(filepath_or_buffer=f"../experiments/scenarioI/results/optuna/CMA-ES_n_shots_{n_shots}_experiments.pkl")
```

```python
df_cobyla.is_target_most_probable.value_counts()
```

```python
mask = df_cobyla.target_probability_after_training > 1/n_shots
df_cobyla[mask].shape
```

```python
df_cmaes.is_target_most_probable.value_counts()
```

```python
mask = df_cmaes.target_probability_after_training > 1/n_shots
df_cmaes[mask].shape
```

```python
df_cmaes.groupby("n_iter").is_target_most_probable.value_counts()
```

```python
mask = df_cmaes.target_probability_after_training > 1/n_shots
df_cmaes[mask].groupby("n_iter").obj_value.count()
```
